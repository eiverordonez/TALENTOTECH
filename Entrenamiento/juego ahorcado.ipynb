```python
import random
# creo una lista de palabras
def elegir_palabra():
    """Elige una palabra aleatoria de una lista."""
    palabras = ["CIBERSEGURIDAD", "phyton", "popayan", "unicauca", "FUNCION", "cauca", "CLASE", "talento", "MODULO", "LIBRERIA","camion"]
    #se elige una palabra al azar
    return random.choice(palabras).upper()

def mostrar_tablero(palabra, letras_correctas):
    """Muestra el tablero del ahorcado."""
    tablero = ""
    for letra in palabra:
        if letra in letras_correctas:
            tablero += letra + " "
        else:
            tablero += "_ "
    print(tablero)

def jugar_ahorcado():
    """Función principal para jugar al ahorcado."""
    palabra_secreta = elegir_palabra()
    letras_correctas = set()
    letras_incorrectas = set()
    intentos_restantes = 7

    print("¡Bienvenido al juego del Ahorcado!")
    print(f"La palabra tiene {len(palabra_secreta)} letras.")

    while intentos_restantes > 0:
        print("\nIntentos restantes:", intentos_restantes)
        mostrar_tablero(palabra_secreta, letras_correctas)
        print("Letras incorrectas:", ", ".join(sorted(list(letras_incorrectas))))

        intento = input("Introduce una letra: ").upper()

        if len(intento) != 1 or not intento.isalpha():
            print("Por favor, introduce una sola letra.")
            continue
        elif intento in letras_correctas or intento in letras_incorrectas:
            print("Ya has intentado esa letra. ¡Intenta otra!")
            continue

        if intento in palabra_secreta:
            letras_correctas.add(intento)
            print("¡Correcto!")
            if set(palabra_secreta) == letras_correctas:
                print("\n¡Felicidades! ¡Has adivinado la palabra:", palabra_secreta + "!")
                break
        else:
            letras_incorrectas.add(intento)
            intentos_restantes -= 1
            print("¡Incorrecto!")

    if intentos_restantes == 0:
        print("\n¡Te has quedado sin intentos!")
        print("La palabra secreta era:", palabra_secreta)

if __name__ == "__main__":
    jugar_ahorcado()
```
