```python
from tabulate import tabulate

def cifrar_vigenere(mensaje, clave):
    """Cifra un mensaje usando el cifrado de Vigenère """
    mensaje = mensaje.upper().replace(' ', '')
    clave = clave.upper()
    resultado = ''
    clave_index = 0
    alfabeto = 'ABCDEFGHIJKLMNÑOPQRSTUVWXYZ'
    n = len(alfabeto)

    for letra_mensaje in mensaje:
        if letra_mensaje in alfabeto:
            indice_mensaje = alfabeto.find(letra_mensaje)
            indice_clave = alfabeto.find(clave[clave_index % len(clave)])
            indice_cifrado = (indice_mensaje + indice_clave) % n
            resultado += alfabeto[indice_cifrado]
            clave_index += 1
        else:
            resultado += letra_mensaje
    return resultado

# Datos frases y claves
datos_cifrado = [
    {"frase": "platzi", "clave": "abc"},
    {"frase": "ciberseguridad", "clave": "utp"},
    {"frase": "talentotech", "clave": "PERSISTE"},
    {"frase": "hola", "clave": "SABIDURIA"},
    {"frase": "SIEMPRE SONRIE ", "clave": "ALEGRIA"},
    {"frase": "popayan ", "clave": "cauca"},
]

tabla_cifrado = []

# Cifrar cada frase y almacenar los resultados para la tabla
for item in datos_cifrado:
    frase_original = item["frase"]
    clave_usada = item["clave"]
    frase_cifrada = cifrar_vigenere(frase_original, clave_usada)
    tabla_cifrado.append([frase_original, clave_usada, frase_cifrada])

# Definir los encabezados de la tabla
encabezados_cifrado = ["Frase Original", "Clave", "Frase Cifrada"]

# Mostrar la tabla de cifrado
print("Tabla de Cifrado Vigenère:")
print(tabulate(tabla_cifrado, headers=encabezados_cifrado, tablefmt="grid"))

def descifrar_vigenere(mensaje_cifrado, clave):
    """Descifra un mensaje cifrado con Vigenère en español."""
    mensaje_cifrado = mensaje_cifrado.upper().replace(' ', '')
    clave = clave.upper()
    resultado = ''
    clave_index = 0
    alfabeto = 'ABCDEFGHIJKLMNÑOPQRSTUVWXYZ'
    n = len(alfabeto)

    for letra_cifrada in mensaje_cifrado:
        if letra_cifrada in alfabeto:
            indice_cifrado = alfabeto.find(letra_cifrada)
            indice_clave = alfabeto.find(clave[clave_index % len(clave)])
            indice_descifrado = (indice_cifrado - indice_clave + n) % n
            resultado += alfabeto[indice_descifrado]
            clave_index += 1
        else:
            resultado += letra_cifrada
    return resultado

tabla_descifrado = []

# Descifrar cada frase cifrada y almacenar los resultados para la tabla
for item in tabla_cifrado:
    frase_cifrada = item[2]
    clave_usada = item[1]
    frase_descifrada = descifrar_vigenere(frase_cifrada, clave_usada)
    tabla_descifrado.append([frase_cifrada, clave_usada, frase_descifrada])

# Definir los encabezados de la tabla de descifrado
encabezados_descifrado = ["Frase Cifrada", "Clave", "Frase Descifrada"]

# Mostrar la tabla de descifrado
print("\nTabla de Descifrado Vigenère:")
print(tabulate(tabla_descifrado, headers=encabezados_descifrado, tablefmt="grid"))
```
